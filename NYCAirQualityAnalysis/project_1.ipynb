### Introduction

#### Dataset Overview
Our analysis utilizes data from the NYC Department of Health and Mental Hygiene's Bureau of Environmental Surveillance and Policy, which details pollutant levels from various locations across New York City. This dataset is crucial for understanding the impacts of air quality on public health and assessing the efficacy of pollution control measures over time.

#### Key Metrics and Terminology
- **NO2 (Nitrogen Dioxide)**: Emitted from combustion processes and contributes to smog and respiratory issues.
- **Ozone (O3)**: A secondary pollutant formed by reaction of sunlight with primary pollutants, harmful at ground level.
- **PM2.5 (Fine Particulate Matter)**: Particles small enough to invade the respiratory system and cause health issues.
- **Air Quality Index (AQI)**: Indicates the level of air pollution and its health implications.
- **NYS DEC SIPs (State Implementation Plans)**: Formulated by the DEC (Department of Environmental Conservation), these plans are designed to reduce air pollution from various sources to meet the NAAQS (National Ambient Air Quality Standards). Targeted pollutants include CO (Carbon Monoxide), Pb (Lead), NO2 (Nitrogen Dioxide), O3 (Ozone), PM2.5 (Particulate Matter 2.5 microns), PM10 (Particulate Matter 10 microns), and SO2 (Sulfur Dioxide).
- **RACT**: (Reasonably Available Control Technology) - A requirement under SIPs for sources to use the most effective emission control technology available.

3. **Seasonal Variability**: Studying how changes in seasons, traffic, and weather affect air quality.

# Seasonal Variations in PM 2.5 Levels in New York City

### Trends Over Time:
- Analysis indicates a general **decrease in PM 2.5 levels** over time, suggesting an improvement in air quality, likely due to stricter emissions regulations and the introduction of cleaner fuel sources for heating and transportation.

### Summer vs. Winter Levels:
- Higher PM 2.5 levels in summer are often due to increased vehicular traffic, including tourist influx and seasonal construction work, compounded by higher temperatures which can accelerate the formation of secondary particulates from existing pollutants.
- In contrast, winter months show elevated PM 2.5 levels from residential heating. However, these have been declining due to the adoption of cleaner heating technologies and regulations limiting the use of heavy heating oils.

### Impact of Traffic Patterns:
- Traffic contributes significantly to PM 2.5 levels, with peaks corresponding to rush hours and seasonal tourism. Initiatives to improve traffic flow and promote public transportation have been effective in reducing these emissions.

### Influence of Weather Conditions:
- Weather conditions such as temperature inversions in winter can trap pollutants near the surface, worsening air quality. Conversely, summer conditions can facilitate the dispersion of pollutants, although this can be offset by the higher production of secondary particulates.

### Effectiveness of Local Policies:
- The downward trend in PM 2.5 levels is a testament to the effectiveness of local policies aimed at reducing emissions from traffic and heating, which are significant contributors to urban air pollution.

### Geographical Factors:
- Geographical features, including the presence of green spaces and proximity to traffic-heavy areas, influence local PM 2.5 levels, highlighting the need for tailored air quality management strategies in different city areas.
